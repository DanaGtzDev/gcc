```python
import matplotlib.pyplot as plt
from PIL import Image

# Columns to visualize
columns = ["origen", "marca", "modelo"]

# Store temporary image paths
image_paths = []

for col in columns:
    plt.figure(figsize=(10, 6))
    
    # Create bar plot
    df_Equipos[col].value_counts().plot(kind="bar")
    
    # Improve formatting
    plt.title(f"Distribución por {col.capitalize()}", fontsize=16)
    plt.xlabel(col.capitalize(), fontsize=12)
    plt.ylabel("Frecuencia", fontsize=12)
    plt.xticks(rotation=45, ha="right")
    plt.grid(axis="y", linestyle="--", alpha=0.4)
    plt.tight_layout()
    
    # Save individual plot
    file_path = f"{col}_barplot.png"
    plt.savefig(file_path, dpi=300)
    plt.close()
    
    image_paths.append(file_path)

# Combine images vertically
images = [Image.open(p) for p in image_paths]

max_width = max(img.width for img in images)
total_height = sum(img.height for img in images)

combined_image = Image.new("RGB", (max_width, total_height))

y_offset = 0
for img in images:
    combined_image.paste(img, (0, y_offset))
    y_offset += img.height

# Save final combined image
final_path = "df_Equipos_barplots_combined.png"
combined_image.save(final_path)

print("Image saved as:", final_path)
```

```python
merged = merged.drop(columns=["description"])
merged = merged.merge(
    df_Refacciones,
    on="order",
    how="left"
)

merged
```

```python
merged = merged.drop(columns=[
    "description",
    "main_plant",
    "g_l_account",
    "material_doc",
    "document_date",
    "material",
    "posting_date",
    "vendor",
    "quantity_in_une"
])

merged
```

```python
merged = merged.rename(columns={
    "quantity": "replacement_quantity",
    "equipment": "equipment_id",
    "marca": "equipment_brand",
    "modelo": "equipment_model",
    "origen": "equipment_origin",
})

merged
```

```python
orders_by_features = (
    merged
    .groupby([
        "plant",
        "equipment_brand",
        "equipment_model",
        "equipment_origin"
    ])["order"]
    .count()
    .reset_index(name="order_count")
)

orders_by_features
```

```python
orders_by_features["percentage"] = (
    orders_by_features["order_count"]
    / orders_by_features["order_count"].sum()
    * 100
)

orders_by_features
```

```python
brand_stats = (
    orders_by_features
    .groupby("equipment_brand")["order_count"]
    .sum()
    .describe()
)

brand_totals = (
    orders_by_features
    .groupby("equipment_brand")["order_count"]
    .sum()
    .sort_values(ascending=False)
)

brand_stats
brand_totals
```
