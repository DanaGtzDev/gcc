```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import MinMaxScaler
from sklearn import metrics
```

```python
sub_df = df[[
    'Used/New',
    'Price',
    'ConsumerRating',
    'ConsumerReviews',
    'SellerType',
    'SellerRating',
    'InteriorColor',
    'Drivetrain',
    'MinMPG',
    'MaxMPG',
    'FuelType',
    'Transmission',
    'Engine',
    'Mileage'
]]

sub_df
```

```python
sub_df.loc[sub_df['Price'].str.contains('Not Priced', case=False, na=False), 'Price'] = '0'
```

```python
sub_df['Price'] = sub_df['Price'].str.replace('$', '', regex=False).str.replace(',','',regex=False)
sub_df['Price']
```

```python
sub_df['Price'] = sub_df['Price'].astype(int)
```

```python
sub_df['Used/New'] = sub_df['Used/New'].where(
    sub_df['Used/New'] == 'Used',
    'New'
)
```

```python
sub_df['Used/New'] = (sub_df['Used/New'] == 'New').astype(int)
```

```python
sub_df = sub_df.drop(columns=["InteriorColor","Drivetrain", "Engine", "Transmission","FuelType"])
```

```python
sub_df = pd.get_dummies(
    sub_df,
    columns=['SellerType'],
    dtype=int
)
```

```python
Y = sub_df["Used/New"]
X = sub_df.drop(columns=["Used/New"])
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    Y,
    test_size=0.2,      # 20% test
    random_state=42     # reproducibility
)
```

```python
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Using 'liblinear' solver and default C=1.0 for demonstration
logreg = LogisticRegression(solver='liblinear', max_iter=1000)
```

```python
# 3. Fit the model
logreg.fit(X_train, y_train)
```

```python
# 4. Predict
y_pred = logreg.predict(X_test)
```

```python
# 5. Evaluate
print("Accuracy:", metrics.accuracy_score(y_test, y_pred))
```
