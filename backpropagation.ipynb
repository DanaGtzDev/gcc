```python
import numpy as np
import random
```

```python
x = [(0,0),(0,1),(1,0),(1,1)]
y = [1,1,1,0]
w1 = random.random()
w2 = random.random()
w3 = random.random()
b1 = 0
b2 = 0
learning_rate = 0.5

epochs = 500
```

```python
def z1(w1,x1,w2,x2,b1):
    return (w1*x1) + (w2*x2) + b1
```

```python
def z2(w3,a,b2):
    return (w3*a) + b2
```

```python
def a(z):
    return 1 / (1 + np.exp(-z))
```

```python
def loss(y,y_pred):
    return 0.5 * np.square(y - y_pred)
```

```python
def delta_1(x,w3,z1,z2,y_pred,y):
    return (x * w3 * np.exp(-z1-z2) * (y_pred - y))/( np.square(1 + np.exp(-z2)) *  np.square(1 + np.exp(-z1))) 
```

```python
def delta_2(a,z2,y_pred,y):
    return (a * np.exp(-z2) * (y_pred - y))/(np.square(1 + np.exp(-z2)))
```

```python
def next_step(w_old, delta):
    return w_old - (learning_rate * delta)
```

```python
def heaviside(x):
    if(x < 0.5):
        return 0
    else:
        return 1
```

```python
for j in range(0, epochs - 1):
    for i in range(0,len(x)):
        z1_value = z1(w1,x[i][0],w2,x[i][1],b1)
        a_value = a(z1_value)
        z2_value = z2(w3,a_value,b2)
        y_pred_value = a(z2_value)
        loss_value = loss(y[i],y_pred_value)

        w1_current, w2_current, w3_current, b1_current, b2_current = w1, w2, w3, b1, b2

        w1 = next_step(w1_current, delta_1(x[i][0],w3_current,z1_value,z2_value,y_pred_value,y[i]))
        w2 = next_step(w2_current, delta_1(x[i][1],w3_current,z1_value,z2_value,y_pred_value,y[i]))
        w3 = next_step(w3_current, delta_2(a_value,z2_value,y_pred_value,y[i]))
        b1 = next_step(b1_current, delta_1(1,w3_current,z1_value,z2_value,y_pred_value,y[i]))
        b2 = next_step(b2_current, delta_2(1,z2_value,y_pred_value,y[i]))

        print(f"Log: w1: {w1}, w2: {w2}, w3: {w3}, b1: {b1}, b2: {b2}, loss:{loss_value}, predicted: {y_pred_value}, real: {y[i]} ")
```
